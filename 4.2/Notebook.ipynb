## Теория

*Бета-распадом* называется самопроизвольное превращение ядер, при котором их массовое число не изменяется, а заряд увеличивается или уменьшается на единицу. Бета-активные ядра встречаются во всей области значений массового числа $A$. Период полураспада $\beta$-активных ядер варьируется от ничтожных долей секнуды до $10^{18}$ лет. 

В данной работе исследуется схема ${}_Z^AX \to {}_{Z+1}^AX + e^{-} + \overline{\nu}$, при которой кроме электрона также испускается антинейтрино. Освобождающаяся при $\beta$ - распаде энергия делится между элекроном, антинейтрино и дочерним ядром, однако доля энергии, передаваемая ядру, крайне мала по сравнению с энергией, уносимой электроном и антинейтрино. Электроны могут иметь любое значение энергии - от нулевой до некоторой максимальной, которая равна энергии, освобождающейся при $\beta$-распаде.

Величина $W(p_e)dp_e$ определяет вероятность того, что частица получит при испускании импульс из интервала $[p_e; p_e+dp_e]$.

В трехмерном пространстве интервалу $[p_e; p_e+dp_e]$ соответствует шаровой слой радиусом $p_e$ и шириной $dp_e$, его объём $4\pi p_e^2dp_e$.

Справедлива формула $W(p_e) dp_e \sim p_e^2p_{\nu}^2dp_e$.

Можно получить $p_{\nu} = \frac {E_{\nu}}c = \frac {T_{max} - T_e}c$, где $E_{\nu}$ - кинетическая энергия антинейтрино, $T_{max}$ - максимально возможная в данном распаде кинетическая энергяи электрона, $T_e$ - его фактическая энергия.

После подстановок можно вывести $W(p_e)dp_e \sim p_e^2(T_{max}-T_e)^2dp_e$.

Данное выражение приводит к спетру, имеющему вид широкого колокола. Дочерние ядра, возникающие в результате $\beta$-распада, нередко оказываются возбужденными. Они отдают свою энергию либо излучая $\gamma$-квант (энергия которого равна разности энергий начального и конечного уровней), либо передавая избыток энергии одному из элеткронов с внутренних оболоечк ядра. Излученные в таком процессе электроны имеют строго определенную энергию и называются конверсионными.

Из инварианта $T_e = \sqrt{p_e^2c^2+m_ec^2}-m_ec^2$ получаем $T_{max}-T_e = c \left (\sqrt{p_{max}^2c^2}-\sqrt{p_e^2+m_e^2c^2} \right )$.

Как показывает расчет, для заряженных частиц тонкая катушка эквивалентна линзе. Её фокусное расстояние $f$ зависит от импульса электронов $p_e$ и от индукции магнитного поля линзы (т.е. от силы тока $I$, протекающего через катушку) так: $\frac 1f \sim \frac {I^2}{p_e^2}$. При этом $p_e=kI$. Константа $k$ обычно определяется из опыта. 

Связь между числом частиц, регистрируемых установкой, и функцией $W(p_e)$ такова: $N(p_e) \sim W(p_e)\Delta p_e$. Здесь $\Delta p_e$ - разрешающая способность спектрометра. 

Дифференцированием формулы для фокуса магнитной линзы при постоянном токе можно получить $\Delta p_e = \frac 12 \frac {\Delta f}{f}p_e$.

Итоговый результат: $N(p_e)= CW(p_e)p_e$, $C$ - некоторая константа.

## Обработка данных

#### Погрешность задания/измерения тока

```python
I_abs_err = 0.01
I_rel_err = 0.05 / 5.0

data['I_err'] = data['I'] * I_rel_err + I_abs_err
```

#### Интенсивность фона

```python
nr_N = data_2['N'].to_numpy()

# Среднее.
normal_radiation = np.mean(nr_N) / period

# Погрешность.
normal_radiation_std = np.std(nr_N) / period
normal_radiation_reg_err = 1.0 / period
normal_radiation_err = np.sqrt(normal_radiation_std**2 + normal_radiation_reg_err**2)

print("Радиационный фон: (%.3f ± %.3f) 1/с" % (normal_radiation, normal_radiation_err))
```

#### Интенсивность излучения

```python
data['intensity'] = (data['N'] / period) - normal_radiation
data['intensity_err'] = np.sqrt((data['N'] / period**2) + normal_radiation_err**2)
```

```python
matplotlib.rc('font', **font)
```

```python
fig, ax = plt.subplots()

ax.set_title("Интенсивность излучения при $\\beta$-распаде в зависимости от тока на катушке")

fig.set_figheight(10)
fig.set_figwidth(16)

#ax.set_xscale('log', basex=10)
#ax.set_yscale('log', basey=10)

ax.grid(color='#000000', alpha=0.15, linestyle='-', linewidth=1, which='major')
ax.grid(color='#000000', alpha=0.1, linestyle='-', linewidth=0.5, which='minor')
#ax.xaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator())
#ax.yaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator())

ax.errorbar(data['I'], data['intensity'], xerr = data['I_err'], yerr = data['intensity_err'],
            ls = '', lw = 1.0, marker = 'x', solid_capstyle='projecting', capsize = 3, label = '2.5В')

#ax.set_xlim(left = 2.0)

ax.set_xlabel('Ток, А')
ax.set_ylabel('Интенсивность, 1/с')

ax.legend(loc='upper left')
plt.show();
```

### Импульсы

```python
max_intensity_idx = data['intensity'].idxmax()

max_intensity = data['intensity'][max_intensity_idx] # Максимальная интенсивность.
max_intensity_err = data['intensity_err'][max_intensity_idx] # Погрешность.
max_intensity_I = data['I'][max_intensity_idx]       # Ток при максимальной интенсивности.
max_intensity_I_err = data['I_err'][max_intensity_idx]

# Границы пика.
I_treshold = 3.5
intensity_relative_treshold = 0.75
max_intensity_Ilb = data[(data['intensity'] > intensity_relative_treshold * max_intensity) & (data['I'] > I_treshold)]['I'].min()
max_intensity_Iub = data[(data['intensity'] > intensity_relative_treshold * max_intensity) & (data['I'] > I_treshold)]['I'].max()

# Ширина пика.
max_intensity_I_width = max_intensity_Iub - max_intensity_Ilb
max_intensity_I_total_err = np.sqrt(max_intensity_I_width**2 + max_intensity_I_err**2)

print("Положение пика: интенсивность - (%.3f ± %.3f) 1/с; ток - (%.3f ± %.3f) А" %
      (max_intensity, max_intensity_err, max_intensity_I, max_intensity_I_total_err))
```

```python
# Константа пропорциональности:
Ip_koeff = T_k / max_intensity_I
Ip_koeff_err = Ip_koeff * (max_intensity_I_total_err / max_intensity_I)

print("Константа пропорциональности между током и импульсом: (%.3e ± %.3e) эВ/А" % (Ip_koeff, Ip_koeff_err))
```

```python
data['p'] = data['I'] * Ip_koeff
data['p_err'] = np.sqrt((Ip_koeff_err * data['I'])**2 + (Ip_koeff * data['I_err'])**2)
```

### График Ферми

```python
data['Fermi'] = np.sqrt(data['intensity'] / (data['p']**3))
data['Fermi_err'] = data['Fermi'] * np.sqrt(0.5 * (data['intensity_err'] / data['intensity'])**2 +
                                            1.5 * (data['p_err'] / data['p'])**2)
```

```python
#data
```

#### Линейная аппроксимация

```python
X_fit_min = 2.5e5
X_fit_max = 5.5e5

X_fit = data[(X_fit_min <= data['p']) & (data['p'] <= X_fit_max)]['p'].to_numpy()
Y_fit = data[(X_fit_min <= data['p']) & (data['p'] <= X_fit_max)]['Fermi'].to_numpy()
```

```python
fitted = np.polyfit(X_fit, Y_fit, deg = 1)
```

```python
x_zero = -fitted[1] / fitted[0]

m_e = 0.511e5
T_max = np.sqrt(x_zero**2 + m_e**2) - m_e
print("Максимальная энергия электронов, эВ: %.3e" % x_zero)

Xf = np.arange(0.0, x_zero + 0.2e5, 100)
Yf = np.array([np.polyval(fitted, x) for x in Xf])
```

```python
p_treshold = 1.0e5
X = data[data['p'] > p_treshold]['p'].to_numpy()
Y = data[data['p'] > p_treshold]['Fermi'].to_numpy()
X_err = data[data['p'] > p_treshold]['p_err'].to_numpy()
Y_err = data[data['p'] > p_treshold]['Fermi_err'].to_numpy()
```
